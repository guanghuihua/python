```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
#matplottlib
# Windows
plt.rcParams['font.sans-serif']=['SimHei'] #
plt.rcParams['axes.unicode_minus']=False  #-2
'''







'''
data = pd.DataFrame(np.random.rand(30,2)*1000,
                   columns = ['A_sale','B_sale'],
                   index = pd.period_range('20170601','20170630'))
print(data.head())
#   30A/B
 
data.plot(kind='line',
       style = '--.',
       alpha = 0.8,
       figsize = (10,3),
       title = 'AB-',
          rot=20)
# 
 
data.plot(kind = 'bar',
          width = 0.8,
          alpha = 0.8,
          figsize = (10,3),
          title = 'AB-')
# 
plt.show()
```

```python
# =================================1  =============
# 3+
 
fig3 = plt.figure(figsize=(10,6))
plt.subplots_adjust(hspace=0.3)
# 
 
ax1 = fig3.add_subplot(2,1,1)
x = range(len(data))
y1 = data['A_sale']
y2 = -data['B_sale']
plt.bar(x,y1,width = 1,facecolor = 'yellowgreen')
plt.bar(x,y2,width = 1,facecolor = 'lightskyblue')
plt.title('AB-')
plt.grid()
plt.xticks(range(0,30,6))
ax1.set_xticklabels(data.index[::6])   #x
# 
ax2 = fig3.add_subplot(2,1,2)
y3 = data['A_sale']-data['B_sale']
plt.plot(x,y3,'--go')
plt.axhline(0,hold=None,color='r',linestyle="--",alpha=0.8)  # y
plt.grid()
plt.title('AB-')
plt.xticks(range(0,30,3))               #   10  
ax2.set_xticklabels(data.index[::3])    #   
# 
plt.show()
```
